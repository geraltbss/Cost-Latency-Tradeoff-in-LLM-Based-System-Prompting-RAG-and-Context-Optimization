```python
import fitz
import os
```

PDF to TXT script

```python
def pdf_to_text(pdf_path, output_path):
    doc = fitz.open(pdf_path)
    full_text = ""

    for page in doc:
        full_text += page.get_text("text") + "\n"

    with open(output_path, "w", encoding="utf-8") as f:
        f.write(full_text)


pdf_to_text("../Dataset/GDPR.pdf", "GDPR.txt")
pdf_to_text("../Dataset/AI Act.pdf", "AIA.txt")
pdf_to_text("../Dataset/DGA.pdf", "DGA.txt")
pdf_to_text("../Dataset/DMA.pdf", "DMA.txt")
pdf_to_text("../Dataset/DSA.pdf", "DSA.txt")
```
